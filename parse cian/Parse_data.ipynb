```python
import urllib.request
from bs4 import BeautifulSoup
```

```python
def get_html(url):
    response = urllib.request.urlopen(url)
    return response.read()
```

## Парсим страницу с квартирой

```python
def parse(html):
    soup = BeautifulSoup(html, "html5lib")
    data = soup.find("div", class_ = "header-information--3XD2V")
    geo = data.find("div", class_ = "geo--1poV5")
    geo = geo.find_all("a")
    st = ""
    for i in geo:
        st += i.text + " " 
    price = soup.find("span", itemprop="price")
    price = price.text.replace(u'\xa0', '').replace(u"₽/мес.", '')
    return st, price
    
```

## Парсим урлы со страницами квартир

```python
def parse_list(html, url_data):
    soup = BeautifulSoup(html, "html5lib")
    data = soup.find_all("a", class_ = "header--28QlS")
    for i in data:
        url_data.append(i['href'])
    return url_data
```

### проходим со страницы 1 по 61 и сохраняем урлы со страницей квартиры

```python
urls = []
for i in range(61)[1:]:
    parse_list(get_html("https://www.cian.ru/cat.php?deal_type=rent&engine_version=2&offer_type=flat&p="+str(i)+"&region=1&room1=1&type=4"), urls)
```

```python
f = open('urls.txt', 'w')
for i in urls:
    f.write('%s\n'%i)
f.close()
```
